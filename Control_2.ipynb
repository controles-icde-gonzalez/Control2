```python
import pandas as pd

# cargar el archivo HDI.xlsx
# lee diferente el encabezado de la columna de la tabla. El encabezado son las filas 4, 5 y 6
excel = pd.read_excel('HDI.xlsx', header = [4, 5, 6]) 

# lee 276 filas, pero la ultima parte de la tabla tiene apartado de 'Notes', 'Definitions', 'Main data sources'
# solo me quedo con las 223 fillas de datos que si tienen los datos de la tabla 
excel = excel.iloc[0:223, :]

# mostrar la tabla
excel
```

```python
# hay columnas que no contienen informacion, pero
# debe salir la informacion final de world que es la ultima fila
excel.iloc[-1]
```

```python
excel.info()
#viendo todas las columnas, hay varias que no contienen informacion de la tabla
```

```python
# hay columnas que no contienen informacion, que son parte de la leyenda, por lo que contienen missing values pero también en algunas filas contienen solo letras
quitar = [('Human Development Index (HDI) ', 'Value', '2021.1'),
        ('Life expectancy at birth', '(years)', '2021.1'),
        ('Expected years of schooling', '(years)', 'a'),
       ('Mean years of schooling', '(years)', 'a'),
         ('Gross national income (GNI) per capita', '(2017 PPP $)', '2021.1'),
      ('GNI per capita rank minus HDI rank', 'Unnamed: 13_level_1', 'b')]
excel = excel.drop(quitar, axis=1)

# algunos datos contienen '..' o '—', remplazar por missing values
excel = excel.replace('..', pd.NA) 
excel = excel.replace('—', pd.NA) 
excel = excel.replace('NaN', pd.NA)

# para quitar espacios que podrian estar adelante y atras de los datos 
# 1. cambiar todas las columnas a object (str) y quitar los espacios (strip)
excel = excel.apply(lambda x: x.astype(str).str.strip())

excel.info()
```

```python
#2. para cambiar todas las columnas menos la de paises a float64
# crear columnas_numericas
columnas_numericas = [('Unnamed: 0_level_0', 'HDI rank', 'Unnamed: 0_level_2'),
    ('Human Development Index (HDI) ', 'Value', 2021),
    ('Life expectancy at birth', '(years)', 2021),
    ('Expected years of schooling', '(years)', 2021),
    ('Mean years of schooling', '(years)', 2021),
    ('Gross national income (GNI) per capita', '(2017 PPP $)', 2021),
    ('GNI per capita rank minus HDI rank', 'Unnamed: 12_level_1', 2021),
    ('HDI rank', 'Unnamed: 14_level_1', 2020)
]
# muestra que todas son de tipo object
excel[columnas_numericas].info()
```

```python
# cambiar las comas ',' por puntos '.' de las [columnas_numericas]
excel[columnas_numericas] = excel[columnas_numericas].apply(lambda x: x.str.replace(',', '.'))

# convertir a numerico (float64)
excel[columnas_numericas] = excel[columnas_numericas].apply(pd.to_numeric, errors='coerce')

# muestra que son de tipo float64
excel[columnas_numericas].info()
```

```python
# la tabla completa solo tiene como columna de tipo object a la que tiene los nombres de paises
excel.info()
```

```python
# ahora que todas las columnas estan segun su tipo, eliminar filas que no contengan datos
# es decir, filas que estan por diseño de la tabla
# para saber que filas contienen las separaciones de human development, analizar las filasde  las columnas numericas 
excel[columnas_numericas]
```

```python
# seleccionar las columnas numéricas
data_col_num = excel[columnas_numericas]

# filtrar filas que son completamente NaN
filas_completamente_nan = data_col_num[data_col_num.isna().all(axis=1)]

filas_completamente_nan
```

```python
# para crear una nueva columna de categorias 'human development' 
excel['Human Development Level'] = ''
excel.loc[0:66, 'Human Development Level'] = ' Very High'
excel.loc[67:116, 'Human Development Level'] = 'High'
excel.loc[117:161, 'Human Development Level'] = 'Medium'
excel.loc[162:194, 'Human Development Level'] = 'Low'
excel.loc[195:, 'Human Development Level'] = 'Other countries or territories'

# volverlo categorico
excel['Human Development Level (order)'] = excel['Human Development Level']
mapper1 = {'Very High': 1, 'High': 2, 'Medium': 3, 'Low': 4, 'Other countries or territories': 5}
excel['Human Development Level (order)'].replace(mapper1, inplace=True)

from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=[1, 2, 3, 4, 5], ordered=True)

excel['Human Development Level'] = excel['Human Development Level'].astype('category')

# quitar la columna de 'Human Development Level (order)' porque no la quiero, solo me quedo con la que dice la cateogria en texto
excel = excel.drop(columns=['Human Development Level (order)']) 

# 'Human Development Level' es la unica de tipo category 
excel.info()
```
